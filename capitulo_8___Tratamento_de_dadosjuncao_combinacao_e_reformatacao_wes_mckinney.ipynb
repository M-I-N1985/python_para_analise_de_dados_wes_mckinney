```python
frame = pd.DataFrame(np.arange(12).reshape((4,3)), 
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]], 
                     columns=[['Ohio', 'Ohio', 'Colorado'], ['Green', 'Red', 'Green']])
frame
```

```python
frame.index.names = ['key1', 'key2']
frame.columns.names = ['estado', 'cor']
frame
```

```python
frame['Ohio']
```

###*Reorganizando e ordenando níveis*

```python
frame.swaplevel('key1', 'key2')  # swaplevel aceita dois números ou nomes de níveis e devolve um novo objeto com os níveis trocados - os dados, porém, permanecem inalterados
```

```python
frame.sort_index(level=1)  # sort_index ordena os dados usando os valores de um só nível
```

```python
frame.sort_index(level=0)
```

```python
frame.swaplevel(0, 1).sort_index(level=0)
```

```python
frame.swaplevel(0, 1).sort_index(level=1)
```

#**Combinando e mesclando conjuntos de dados**


*   *pandas.merge conecta linhas em DataFrames com base em um aou mais chaves. Essa operação será conhecida dos usuários de SQL ou de outros bancos de dados relacionais, pois ela implementa as operações de junção (join) dos bancos de dados.*

*   *pandas.concat concatena ou “empilha” objetos ao longo de um eixo.*

###**Junções no DataFrame no estilo de bancos de dados**

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'dados1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'dados2': range(3)})
df2
```

```python
pd.merge(df1, df2)  # uma junção de muitos-para-um (many-toone)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],  'dados1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'dados2': range(3)})
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

*valores 'c' e 'd' e os dados associados estão ausentes no resultado. Por padrão, merge executa
uma junção do tipo 'inner' (interna); as chaves no resultado são a **intersecção**, ou o conjunto comum que se encontra nas duas tabelas. Outras opções possíveis são: ‘left', 'right' e 'outer'. A junção externa (outer join) efetua a união das chaves, combinando o efeito da aplicação das junções tanto à esquerda quanto à direita*

```python
pd.merge(df1, df2, how='outer')
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'], 'dados1': range(6)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'], 'dados2': range(5)})
df2
```

```python
pd.merge(df1, df2, on='key', how='left')
```

*Junções de muitos para muitos formam o produto cartesiano das linhas. Como havia três linhas 'b' no DataFrame da esquerda e duas no da direita, há seis linhas 'b' no resultado. O método de junção afeta somente os valores de chaves distintos que aparecem no resultado*

```python
pd.merge(df1, df2, how='inner')
```

```python
esquerda = pd.DataFrame({'key1': ['foo', 'foo', 'bar'], 'key2': ['one', 'two', 'one'], 'lval': [1, 2, 3]})  # Para um merge com várias chaves, passe uma lista de nomes de coluna
esquerda
```

```python
direita = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'], 'key2': ['one', 'one', 'one', 'two'], 'rval': [4, 5, 6, 7]})
direita
```

```python
pd.merge(esquerda, direita, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(esquerda, direita, on='key1')
```

```python
pd.merge(esquerda, direita, on='key1', suffixes=('_esquerda', '_direita'))
```
